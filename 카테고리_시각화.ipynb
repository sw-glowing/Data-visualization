```python
# dictionary 컬럼에서 품사 정보만 추출
df['pos'] = df['dictionary'].apply(lambda x: x.split(' ')[0])

# 품사별로 개수를 세고, 비율을 구하여 시각화
pos_counts = df['pos'].value_counts()
pos_counts.plot(kind='bar', title='품사별 비율')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
sns.countplot(x='category', data=df, palette='viridis')
plt.title('카테고리별 비율')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
sns.countplot(x='pos', data=df, palette='viridis')
plt.title('품사별 비율')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
df['category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('카테고리별 비율')

plt.subplot(1, 2, 2)
df['pos'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('품사별 비율')

plt.tight_layout()
plt.show()
```

```python
category_pos_counts = df.groupby(['category', 'pos']).size().unstack()

category_pos_counts.plot(kind='bar', stacked=True)
plt.title('카테고리별 품사 분포')
plt.show()
```

```python
colors = ["#F94B6C", "#FF9640", "#A6D146"]
plt.figure(figsize=(12, 5))

# Pie chart
plt.subplot(1, 2, 1)
df['category'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=colors, wedgeprops={'edgecolor': 'white'})
plt.title('Category Distribution', fontsize=14)
plt.ylabel('')

# Stacked bar chart
plt.subplot(1, 2, 1)
category_pos_counts = df.groupby(['category', 'pos']).size().unstack()
category_pos_counts.plot(kind='bar', stacked=True, color=colors, edgecolor='white')
plt.title('POS Distribution by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='POS', title_fontsize=10, fontsize=10)

plt.tight_layout()
plt.show()
```
